- step:

    - lopk about step sign is it positive or negative sign
 
    - positive means increment
 
        - last:stop-1
 
    -  negative means decrement
 
        - last:stop+1  

```python
#example 1:range(3,20,2
#start=3
#step=+2 positive direction
#last=20-1=19
#3 5 7 9 11 13 15 17
for i in range(3,19,2):
    print(i,end=' ')
```

```python
#Q9) ask the user get 5 random numbers #means you need a gate a random number inside for loop
#the loop should run 5 times
#perform the even and odd operation
#1)even count 2)sum of even number
import random
summ_even=0
summ_odd=0
total_even_counter=0
total_odd_counter=0
for  i in range(5):
    a=random.randint(1,100)
    if a%2==0:
        summ_even=summ_even+a
        total_even_counter=total_even_counter+1
        print("even:",a)
    else:
        summ_odd=summ_odd+a
        total_odd_counter=total_odd_counter+1
        print("odd:",a)
print("The sum of even numbers is:",summ_even)
print("The sum of odd numbers is:",summ_odd)
print("the count of even number is:",total_even_counter)
print("the count of odd number is:",total_odd_counter)
     
```
